# Set of genres

We will now see if we can create a set of all genres that are in the list of movies. This can easily be done with a single comprehension if you use two for-loops. Let's work it out in a few simple steps. First, We will look at an example.

# Example

```python
sentence = "there is no place like home"
```

```python
words = sentence.split()
```

```python
words
```

In the comprehension below, we use two for-loops. The first loop iterates over the elements in `words`, assigning one elemner to `w`. The second for-loop iterates over the letters in a single word `w`, assigning every letter to `c`. The expression is the letter `c`, which is put in a set comprehension. In other words, duplicate letters are automatically ignored. The end result is a set of all letters that appear in one or more of the `words`. 

```python
{ c for w in words for c in w }
```
